```python
import logging
import sqlanydb
import datetime
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import pandas as pd
import seaborn as sns

# global database connection
_conn = None
DB_SERVERNAME="dbnb"
DB_NAME="dbnb"
DB_FILE="/home/tom/src/airbnb/db/dbnb.db"
PIECHART_EXPLODE = 0.05
PLOT_TYPE_BAR = 'bar'
PLOT_TYPE_PIE = 'pie'
PLOT_TYPE_SCATTER = 'scatter'

logger = logging.getLogger()

def connect():
    try:
        global _conn
        if _conn is None:
            _conn = sqlanydb.connect(
                userid="dba",
                password="sql",
                serverName=DB_SERVERNAME,
                databasename=DB_NAME,
                databasefile=DB_FILE)
        return _conn
    except:
        logger.error(
            "Failed to connect to database." +
            "You may need to change the DB_FILE value.")

conn = connect()

# basic plotting functions
    
def bar_plot(result_set, title, xlabel, ylabel, scale):
    sns.set(rc={"figure.figsize": (8, 6)})
    sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
    labels = [x for x, y in result_set]
    total = sum([y for x,y in result_set])
    if scale=='percent':
        y = [(y * 100.0)/total for x, y in result_set]
    else:
        y = [y for x, y in result_set]
    x = [labels.index(x) for x, y in result_set]
    #y = [y for x, y in result_set]
    (width, opacity) = (0.4, 0.5)
    plt.xlabel(xlabel)
    if scale=="percent":
        plt.ylabel(ylabel + " (%)")
    else:
        plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(x, labels)
    rects = plt.bar(x, y, align='center', width=width, alpha=opacity)
    # compute heights as a percent
    for rect in rects:
        height = rect.get_height()
        vertical_position = 0.9 * height
        plt.text(rect.get_x()+rect.get_width()/2., 
                 vertical_position, '{0:1.0f}'.format(height), 
                 ha='center', va='bottom')
    plt.show()  
    
def piechart_plot(result_set, title):
    #sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
    sns.set(rc={"figure.figsize": (6, 6)})
    (labels, fractions,) = ([x for x, y in result_set],
                                 [float(y) for x, y in result_set])
    # explode = [PIECHART_EXPLODE if y < 25 else 0.0 for y in fractions]
    explode = [PIECHART_EXPLODE for y in fractions]
    plt.title(title, fontsize='x-large', fontweight='bold')
    patches, texts, autotexts = plt.pie(fractions,
                                        labels=labels,
                                        explode=explode,
                                        colors = sns.color_palette('muted'),
                                        shadow=False,
                                        autopct='%1.1f',
                                        )
    font_properties = fm.FontProperties()
    font_properties.set_size('x-large')
    plt.setp(autotexts, fontproperties=font_properties)
    plt.setp(texts, fontproperties=font_properties)
    plt.show()
    
def scatter_plot(df, title, labels=False):
    for index, row in df.iterrows():
        if labels:
            plt.annotate(
                row[2], 
                xy = (row[0], row[1]), xytext = (-20,20),
                textcoords = 'offset points', ha = 'right', va = 'bottom',
                #bbox = dict(boxstyle = 'round,pad=0.5', alpha = 0.5),
                arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0')
            )
    sns.set(rc={"figure.figsize": (8, 6)})
    columns = list(df.columns.values)
    sns.regplot(columns[0],columns[1], df, label="neighborhoods",
            scatter_kws={"s": 100, "alpha": 1})
    ax = plt.axes()
    ax.set_xlim(0,)
    ax.set_ylim(0,)
    plt.show()

# utility functions
def survey_from_search_area(search_area):
    cur = conn.cursor()
    sql = """select top 1 survey_description, survey_id 
    from survey key join search_area 
    where search_area.name = ?
    and survey_date < today()
    order by survey_id desc"""
    cur.execute(sql, (search_area,))
    [title, survey_id] = cur.fetchone()
    cur.close()
    return [title, survey_id]

# plot types

def room_type_plot(search_area, by_what, plot_type, scale):
    [title, survey_id] = survey_from_search_area(search_area)
    title = title + ": " + by_what
    cur = conn.cursor()
    sqlpre = "select room_type, "
    sqlpost = """ from room 
    where survey_id = ? 
    and room_type is not Null 
    group by room_type
    order by room_type"""
    if by_what == 'rooms' or by_what == 'listings':
        by_what_insert = 'count(*)'
    elif by_what == 'visits':
        by_what_insert = 'sum(reviews)'
    elif by_what == 'income':
        by_what_insert = 'sum(reviews * price)'
    elif by_what == 'price':
        by_what_insert = 'avg(price)'
    elif by_what == 'rating':
        by_what_insert = 'avg(overall_satisfaction)'
    else:
        by_what_insert = 'count(*)'   
    sql = sqlpre + by_what_insert + sqlpost
    cur.execute(sql, (survey_id,))
    rs = cur.fetchall()
    cur.close()
    (xlabel, ylabel)  = ('Room Type', by_what)
    if plot_type == PLOT_TYPE_BAR:
        bar_plot(rs, title, xlabel, ylabel, scale)
    elif plot_type == PLOT_TYPE_PIE:
        piechart_plot(rs, title) 
    else:
        # pie by default
        piechart_plot(rs, title) 
    
def nightly_rate_plot(search_area, by_what):
    [title, survey_id] = survey_from_search_area(search_area)
    cur = conn.cursor()
    sqlpre = "select sum(reviews) review_count, "
    sqlpost = " from room where survey_id = ? and room_type is not Null group by room_type"
    if by_what == 'rooms' or by_what == 'listings':
        by_what_insert = 'count(*)'
    elif by_what == 'visits':
        by_what_insert = 'sum(reviews)'
    elif by_what == 'income':
        by_what_insert = 'sum(reviews * price)'
    else:
        by_what_insert = 'count(*)'   
    sql = sqlpre + by_what_insert + sqlpost
    cur.execute(sql, (survey_id,))
    rs = cur.fetchall()
    cur.close()
    (xlabel, ylabel)  = ('Room Type', 'Number of Listings')
    bar_plot(rs, title, xlabel, ylabel)
    #piechart_plot(rs, title) 
    
def multilister_plot(search_area, by_what, scale):
    [title, survey_id] = survey_from_search_area(search_area)
    title += ": " + by_what
    cur = conn.cursor()
    sqlpre = "select if multilister = 1 then 'multi' else 'single' endif, "
    sqlpost = " from host where survey_id = ? group by multilister order by multilister"
    if by_what == 'hosts':
        by_what_insert = 'count(*)'
    elif by_what == 'listings':
        by_what_insert = 'sum(rooms)'
    elif by_what == 'visits':
        by_what_insert = 'sum(review_count)'
    else:
        by_what_insert = 'sum(income1)'   
    sql = sqlpre + by_what_insert + sqlpost
    cur.execute(sql, (survey_id,))
    rs = cur.fetchall()
    cur.close()
    (xlabel, ylabel)  = ('Multilister', by_what)
    bar_plot(rs, title, xlabel, ylabel, scale)    
    #piechart_plot(rs, title)    
```

# Evolution

```python
# plot function.
def date_line_plot(result_set_list, label, ymin, ymax, legend_labels):
    sns.set(rc={"figure.figsize": (8, 6)})
    sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
    #sns.set_palette("GnBu_d",len(result_set_list))
    fig, ax = plt.subplots()
    ax.set_xlim(datetime.datetime.strptime("2013-11-01", "%Y-%m-%d" ), datetime.datetime.today())
    if ymin!=None and ymax!= None:
        ax.set_ylim(ymin, ymax)
    fig.autofmt_xdate()
    plt.xlabel('Date')
    plt.ylabel(str.capitalize(label))
    plt.title(str.title(label))
    for result_set in result_set_list:
        cols = len(result_set[0]) - 1 # each row is assumed to be [x, y1, y2, ...]
        y = []
        x = [datetime.datetime.strptime( row[0], "%Y-%m-%d" ) for row in result_set]
        for col in range(cols):
            y.append([row[col + 1] for row in result_set])
        for col in range(cols):
            plt.plot_date(x, y[col], '-')
            legend = ax.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.show()  
```

```python
# query execution
def evolution(property, search_area_list):
    rs = []
    if property == "rating":
        (ymin, ymax) = (4.0, 5.0)
        label = "average rating"
        sql_ins = "avg(overall_satisfaction)"
    elif property == "listing":
        (ymin, ymax) = (None, None)
        label = "number of listings"
        sql_ins = "count(*)"
    elif property == "review":
        (ymin, ymax) = (None, None)
        label = "number of reviews"
        sql_ins = "avg(reviews)"
    elif property == "price":
        (ymin, ymax) = (None, None)
        label = "price per night"
        sql_ins = "avg(price)"
    elif property == "accommodates":
        (ymin, ymax) = (None, None)
        label = "accommodates"
        sql_ins = "avg(accommodates)"
    for search_area in search_area_list:
        cur = conn.cursor()
        sql_pre = "select s.survey_date, "
        sql_post = """ from room r 
        key join survey s 
        key join search_area sa
        where sa.name = ?
        and deleted = 0
        group by s.survey_date
        order by s.survey_date"""
        sql = sql_pre + sql_ins + sql_post
        cur.execute(sql, (search_area,))
        rs.append(cur.fetchall())
        cur.close()
    date_line_plot(rs, label, ymin, ymax, search_area_list)
```

```python
# arg0 = 'listing' or 'rating' or 'review' or 'price' or 'accommodates'
# arg1 = list of cities
evolution("listing", 
          ['Paris', 'Rome', 'Amsterdam', 'London', 'New York', 'Boston', 'Los Angeles', 'Montreal',
           'San Francisco', 'Berlin', 'Barcelona', 'Sydney', 'Rio de Janeiro'])
```

# Exit and Entry

Look at the ratings of rooms that leave the site and those that come onto the site. If the rating system eliminates bad actors then one would expect the exiters to have lower ratings than those who stay on the site.

These plots show ratings (or other properties) for cities between their earliest and latest surveys on the site. The "stayed" value is computed at the earlier survey for comparison with the "exited".

```python
def exit_plot(property, city_list):
    plot_count = len(city_list)
    axes_tuple = (0) * plot_count
    (width, opacity) = (0.3, 0.5)
    fig, axes_tuple = plt.subplots(1, plot_count, figsize=(12, 3)) # rows, cols
    i_axes = 0
    
    #sns.set(rc={"figure.figsize": (8, 6)})
    #sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 2.5})
    
    for city in city_list:
        ax = axes_tuple[i_axes]
        cur = conn.cursor()
        sql = """
        select min(survey_id), max(survey_id) 
        from survey key join search_area sa 
        where sa.name = ?
        """
        cur.execute(sql, (city,))
        (old, new) = cur.fetchall()[0]
        sql_pre = "select Category, "
        sql_post = " from survival_indicators(?,?)"
        if property=="rating":
            sql_ins = "mean_rating"
        elif property == "number":
            sql_ins = "number"
        elif property == "review":
            sql_ins = "mean_reviews"
        sql = sql_pre + sql_ins + sql_post
        cur.execute(sql, (old, new,))
        result_set = cur.fetchall()
        cur.close()
        categories = [row[0] for row in result_set]
        ratings =  [row[1] for row in result_set]
        xticks = [categories.index(category) + 0.5 for category in categories]
        ax.set_xlabel('Category')
        ax.set_ylabel(str.capitalize(property))
        ax.bar(xticks, ratings, align='center', width=width, alpha=opacity);
        ax.set_xlim([0, len(categories)])
        ax.set_ylim([4, 1.1 * max(ratings)])
        ax.set_xticks(xticks)
        ax.set_xticklabels(categories, rotation=45)
        ax.set_title(city)
        i_axes += 1
    plt.setp(fig.axes, xlabel="Category")
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.show()  

# arg0 = 'rating' or 'number' or 'review'
# arg1 = city list
exit_plot('number',
        ['Amsterdam', 'Paris',  'New York', 'Rome'])
```

# Berlin City Report

This chart compares estimated income per neighborhood from my data (x axis) with the actual reported income per neighborhood from Airbnb's Berlin report. The overall good fit of the graph shows that the estimated income provides a good overall picture of the actual income from Airbnb at the neighborhood level.

```python
# Airbnb Berlin Report
cur = conn.cursor()
sql = """
select  sum(reviews*price)/100 est_income,
        floor(earnings * 1000) income, 
        rh.neighborhood nbhd
from airbnb.report_berlin rb 
join room_history(3) rh
on rb.neighborhood = rh.neighborhood
group by nbhd, income
order by income desc
"""
cur.execute(sql, )
df = pd.DataFrame(data=cur.fetchall(), 
                  columns=["Estimated income", "Income", "neighborhoods", ])
cur.close()
scatter_plot(df, 'Berlin', True)
```

# Ratings

The distribution of ratings in Airbnb listings, from city to city

```python
def rating_plot(search_area_list):
    plot_count = len(search_area_list)
    axes_tuple = (0) * plot_count
    (width, opacity) = (0.3, 0.5)
    fig, axes_tuple = plt.subplots(1, plot_count, figsize=(12, 3)) # rows, cols
    i_axes = 0
    
    for search_area in search_area_list:
        [title, survey_id] = survey_from_search_area(search_area)
        sql = """select r.overall_satisfaction rating, count(*) n 
        from room r key join survey s
        where s.survey_id = ?
        and price is not null
        group by rating
        having rating is not null"""
        cur = conn.cursor()
        cur.execute(sql, (survey_id,))
        rs = cur.fetchall()
        cur.close()
        rating = [row[0] for row in rs]
        count = [int(row[1]) for row in rs]
        ax = axes_tuple[i_axes]
        ax.set_ylabel('Number of Listings')
        ax.set_xlim([0,5.5])
        ax.set_ylim([0,1.1 * max(count)])
        ax.set_title(title)
        ax.bar(rating, count, align='center', alpha=opacity, width=width);
        i_axes += 1
    plt.setp(fig.axes, xlabel="Rating")
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
    plt.show()

rating_plot(['Amsterdam', 'Paris', 'New York', 'Washington'])
```
